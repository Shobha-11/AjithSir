```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
my_label=["Apple", "Mango", "Kiwi", "Orange"]
plt.pie(x, labels=my_label)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
my_label=["Apple", "Mango", "Kiwi", "Orange"]
my_explode=[0.2,0,0,0]
plt.pie(x, labels=my_label, explode=my_explode)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('mela.csv')
df.plot(color=['red','blue','brown'])
plt.title('Mela Sale Report')
plt.xlabel('Days')
plt.ylabel('Sales in Rs')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('mela.csv')
df.plot(color=['red','blue','brown'], marker="*", markersize=10, linestyle="--")
plt.title('Mela Sale Report')
plt.xlabel('Days')
plt.ylabel('Sales in Rs')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('mela.csv')
df.plot(kind='bar', x='Day', title="Mela Sales Report")
plt.ylabel('Sales in Rs')
plt.show()
df
```
